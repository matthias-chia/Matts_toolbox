```python
import pyautogui as ag
#initialise the key/click worker
pyautogui.PAUSE = 1.5
pyautogui.FAILSAFE = True
#find screensize
#pyautogui.size()
width, height = pyautogui.size()
print(width,height)
```

```python
import pyautogui as ag
#initialise the key/click worker
ag.PAUSE = 1.5
ag.FAILSAFE = True

##Move mouse
for i in range(2):
    ag.moveTo(100, 100, duration=0.25)
    ag.moveTo(200, 100, duration=0)
    ag.moveTo(200, 200, duration=0.25)
    ag.moveTo(100, 200, duration=0)
```
