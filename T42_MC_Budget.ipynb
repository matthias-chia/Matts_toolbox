### Monte Carlo simulation: Forecasting Firm future gross profit 

$\textbf{Methodology}$

This is a basic single step markov chain for a static normal probability distribution for $x_{t+1}$ using the conditional probability of $x_{t}$

$$
\operatorname{E} (x_{t+1} \mid x_{t})
$$ 

$\textbf{Data}$

Profit = Revenue - Cost of Goods Sold(COGS)

#### Import packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Import Model (Historical)

```python
#normal 
rMean=170 #mean revenue (in millions $)
rSD=20 #(in millions $)
iterations=10000 #define number of iterations (distribution to be programmed later)
```

#### Generate n-iterations of revenue for chosen distribution

```python
rev = np.random.normal(rMean, rSD,iterations)
rev
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

#### Generate n-iterations of Cost of Goods Sold (COGS) for chosen distribution

```python
#historical
COGSMean = 0.6
COGSsd = 0.1
#generate COGS
COGS = -(rev*np.random.normal(COGSMean,COGSsd))
plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show
```

#### Gross Profit (GP)

```python
gProfit = rev + COGS
gProfit

plt.figure(figsize=(15,6))
plt.plot(gProfit)
plt.xlabel('Iterations')
plt.ylabel('Gross Profit - Simulated')
plt.show
```

```python
print('max profit ($m):', round(max(gProfit),2))
print('min profit ($m):', round(min(gProfit),2))
print('mean profit($m):', round(gProfit.mean(),2))
print('std profit ($m):', round(gProfit.std(),2))
```

#### Plot Histogram (showing distribution)

```python
#Define bin size
plt.figure(figsize=(15,6))
plt.hist(gProfit, bins = [50,60,70,80,90,100,110,120])
plt.xlabel("Monte-Carlo Sim: Gross Profit, n ="+ str(iterations))
plt.ylabel('Count')
plt.show
```
